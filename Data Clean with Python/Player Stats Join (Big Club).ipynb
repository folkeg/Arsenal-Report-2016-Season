```python
import pandas as pd
import xlrd
```

```python
xl = pd.ExcelFile("../Raw Data/Player Stats/Club Stats.xlsx")
goals = xl.parse("Goals")
assists = xl.parse("Assists")
passes = xl.parse("Passes")
shoots = xl.parse("Shots")
tackles = xl.parse("Tackles")
touches = xl.parse("Touches")
clearence = xl.parse("Clearences")
big_chance_missed = xl.parse("Big Chance Missed")
xl = pd.ExcelFile("../Raw Data/Player Stats/PlayerPosition.xlsx")
position = xl.parse("Position")
xl = pd.ExcelFile("../Raw Data/Player Stats/Player Value 2016.xlsx")
value = xl.parse("Value")
```

```python
result = pd.merge(touches, position, how='left', on='Player')
result
```

```python
result = pd.merge(assists, result, how='right', on=['Club', 'Player'])
result = pd.merge(shoots, result, how='right', on=['Club', 'Player'])
result = pd.merge(goals, result, how='right', on=['Club', 'Player'])
result = pd.merge(tackles, result, how='right', on=['Club', 'Player'])
result = pd.merge(clearence, result, how='right', on=['Club', 'Player'])
result = pd.merge(big_chance_missed, result, how='right', on=['Club', 'Player'])
result = pd.merge(passes, result, how='right', on=['Club', 'Player'])
result
```

```python
result = pd.merge(value, result, how='right', on='Player')
result
```

```python
result = result.fillna(value = 0)
result
```

```python
result.to_csv('../Cleaned Data/EPL Player Stats(Big Club).csv', encoding='utf-8')
```
