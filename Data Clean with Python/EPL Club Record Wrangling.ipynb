# EPL Team Data From ESPN

```python
import pandas as pd
from bs4 import BeautifulSoup
import bs4
import requests
```

```python
team_dict = {}
df = pd.DataFrame()
year = 2016

url_init = 'http://www.espnfc.com/barclays-premier-league/23/table'

for i in range(0, 16):
    if i == 0:
        url = url_init
    else:
        url = url_init + "?season=" + (str)(year - i)
    r  = requests.get(url)
    soup = BeautifulSoup(r.text, 'html5lib') 
    for tr in soup.find_all('tr', attrs={'style':True}):
        j = 0
        for td in tr.find_all('td'):
            if j == 1:
                team_dict[j] = td.get_text().strip('\n\t')
            elif j == 2:
                team_dict[j] = year - i
            else:
                team_dict[j] = td.get_text()
            j += 1
        df = df.append(team_dict, ignore_index=True)

cols = ['Pos', 'Team','Season','Games','W_ov','D_ov','L_ov','G_ov','A_ov','Empty',
            'W_hm','D_hm','L_hm','G_hm','A_hm', 'Empty','W_aw',
            'D_aw','L_aw','G_aw','A_aw','Empty','GD','PTS']
df.columns = cols

df.head()
```

```python
df['Season'] = df['Season'].astype(int)
df = df.drop('Empty', 1)
df
```
