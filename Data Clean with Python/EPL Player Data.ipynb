```python
import pandas as pd
import xlrd
```

```python
xl = pd.ExcelFile("../Raw Data/Player Stats/Minutes.xlsx")
minutes = xl.parse("Minutes")
xl = pd.ExcelFile("../Raw Data/Player Stats/Assist.xlsx")
assists = xl.parse("Assist")
xl = pd.ExcelFile("../Raw Data/Player Stats/Passes.xlsx")
passes = xl.parse("Passes")
xl = pd.ExcelFile("../Raw Data/Player Stats/Goal.xlsx")
goals = xl.parse("Goal")
xl = pd.ExcelFile("../Raw Data/Player Stats/Shoots.xlsx")
shoots = xl.parse("Shoots")
xl = pd.ExcelFile("../Raw Data/Player Stats/Tackle.xlsx")
tackles = xl.parse("Tackle")
xl = pd.ExcelFile("../Raw Data/Player Stats/Touch.xlsx")
touches = xl.parse("Touch")
xl = pd.ExcelFile("../Raw Data/Player Stats/Clearence.xlsx")
clearence = xl.parse("Clearence")
xl = pd.ExcelFile("../Raw Data/Player Stats/CleanSheet.xlsx")
cleansheet = xl.parse("CleanSheet")
xl = pd.ExcelFile("../Raw Data/Player Stats/PlayerPosition.xlsx")
position = xl.parse("Position")
xl = pd.ExcelFile("../Raw Data/Player Stats/Big Chance Missed.xlsx")
big_chance_missed = xl.parse("Big Chance Missed")
value = pd.read_csv("../Raw Data/Player Stats/Player Value.csv")
```

```python
result = pd.merge(touches, minutes, how='outer', on=['Season', 'Player'])
result
```

```python
result = pd.merge(assists, result, how='right', on=['Season', 'Player'])
result = pd.merge(shoots, result, how='right', on=['Season', 'Player'])
result = pd.merge(goals, result, how='right', on=['Season', 'Player'])
result = pd.merge(tackles, result, how='right', on=['Season', 'Player'])
result = pd.merge(cleansheet, result, how='right', on=['Season', 'Player'])
result = pd.merge(clearence, result, how='right', on=['Season', 'Player'])
result = pd.merge(big_chance_missed, result, how='right', on=['Season', 'Player'])
result = pd.merge(passes, result, how='right', on=['Season', 'Player'])
result = pd.merge(position, result, how='right', on='Player')
result
```

```python
value
value = value.drop("Unnamed: 0", axis =1 )
```

```python
result = pd.merge(value, result, how='right', on=['Player','Season'])
result
```

```python
result = result.fillna(value = 0)
```

```python
result.to_csv('../Cleaned Data/EPL Player Stats.csv', encoding='utf-8')
```
