# Player Value

```python
import pandas as pd
from bs4 import BeautifulSoup
import bs4
import requests
```

```python
player_dict = {}
df = pd.DataFrame()
 
team = ['Man City', 'Man Utd', 'Arsenal','Tottenham Hotspur','Liverpool']
year = 2016

for i in range(0, 5):
    for j in range(0, 7):
        soup = BeautifulSoup(open('../Raw Data/Player Market Value Html/' + team[i] + ' ' + str(year - j) + ' Value.html'), 'lxml')
        for tr in soup.find('table', attrs = {'class':'items'}).find_all('tr', attrs = {'class':['odd','even']}):
            player_dict[0] = year - j
            if tr.find_all('td')[1].find('a', {'class':'spielprofil_tooltip'}):
                player_dict[1] = tr.find_all('td')[1].find('a', {'class':'spielprofil_tooltip'}).get_text()
            player_dict[2] = tr.find_all('td')[-1].get_text()
            df = df.append(player_dict, ignore_index=True)
    
cols = ['Season','Player','Market_Value']
df.columns = cols

df

```

```python
df.Market_Value = df.Market_Value.str.rstrip().str[1:-1]
df
```

```python
df['Season'] = df['Season'].astype(int)
df.to_csv('../Cleaned Data/Player Value.csv', encoding='utf-8')
```
