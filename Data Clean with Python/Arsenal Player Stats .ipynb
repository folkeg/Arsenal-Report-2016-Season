```python
import pandas as pd
import xlrd
```

```python
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Minutes.xlsx")
minutes = xl.parse("Minutes")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Weekly Wage.xlsx")
wage = xl.parse("Weekly Wage")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Assists.xlsx")
assists = xl.parse("Assists")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Passes.xlsx")
passes = xl.parse("Passes")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Goals.xlsx")
goals = xl.parse("Goals")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Shoots.xlsx")
shoots = xl.parse("Shoots")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Tackles.xlsx")
tackles = xl.parse("Tackles")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Touches.xlsx")
touches = xl.parse("Touches")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Clearence.xlsx")
clearence = xl.parse("Clearence")
xl = pd.ExcelFile("../Raw Data/Player Stats/Arsenal-Position.xlsx")
position = xl.parse("Position")
```

```python
result = pd.merge(wage, minutes, how='right', on=['Season', 'Player'])
result
```

```python
result = pd.merge(assists, result, how='outer', on=['Season', 'Player'])
result = pd.merge(passes, result, how='outer', on=['Season', 'Player'])
result = pd.merge(shoots, result, how='outer', on=['Season', 'Player'])
result = pd.merge(goals, result, how='outer', on=['Season', 'Player'])
result = pd.merge(tackles, result, how='outer', on=['Season', 'Player'])
result = pd.merge(touches, result, how='outer', on=['Season', 'Player'])
result = pd.merge(clearence, result, how='outer', on=['Season', 'Player'])
result
```

```python
result = pd.merge(position, result, how='right', on='Player')
result
```

```python
result = result.fillna(value = 0)
result
```

```python
result.to_csv('../Cleaned Data/Arsenal Stats.csv', encoding='utf-8')
```
